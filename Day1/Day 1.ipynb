# 3. Handling Missing data

Why to handle missing data in ML model?-- the performace of the model decreases, the preditive capabilities decrease and may lead to wrong results..


What to do--? 
1. We can replace the categorical variables with thier mode and continuous variables with their mean values(Generally this is done in the machine learning models).
2. We can replace the null values with 0
3. We can drop the rows which contain null values(generally this is not done)



Lets try identifying and solving the null value problem in the above dataset

```python
ds.head()
```

```python
ds.shape
```

```python
ds.isnull()
```

```python
ds.isnull().sum()
```

Age and salary have one null value each respectively

# Replacing null values with zero

```python
ds_clean1=ds.fillna(value='0')
```

```python
ds_clean1.isnull().sum()
```

ds_clean1 does not have null values as they are all replaced by 0 by fillna function

# Filling null values with the preceeding value in the dataset(Row wise)

```python
ds_clean2=ds.fillna(method='pad')
```

```python
ds
```

```python
ds_clean2
```

Row 4 now has 61000 instead of NaN

# Filling null values with the succesive value in the dataset (Rowwise)

```python
ds_clean3=ds.fillna(method='bfill')
```

```python
ds_clean3
```

4th row has 58000 same as row5 instead of NaN in salary column.

# Filling null values with the preceeding value in the dataset(Columnwise)

```python
ds_clean4=ds.fillna(method='pad', axis=1)
ds_clean4
```

# Filling null values with the succesive value in the dataset(Columnwise)

```python
ds_clean5=ds.fillna(method='bfill',axis=1)
```

```python
ds_clean5
```

# Different null values to be replaced with different values

```python
ds_clean6=ds.fillna({'Salary': '0', 'Age' : '10'}) 
```

```python
ds_clean6
```

# Replacing Null values with mean/median/max/min

```python
ds_clean7=ds.fillna(value=round(ds['Salary'].mean(),2))
```

```python
ds_clean7
```

4th row null is replaced by the salary column's mean. Null values in continuous variables are generally replaced by their mean values in ml models.

# Dropping rows with null values- dropna() function

```python
ds_clean8=ds.dropna()
```

```python
ds_clean8
```

Two rows 4,6 which had just one null value each resulted in the whole row to be deleted, if we want the row to be deleted if all the values are null we can use how parameter inside dropna 

```python
ds_clean9=ds.dropna(how='all')
```

```python
ds_clean9
```

None of the rows were deleted in this case as all the row values were not missing.

# Using replace function to replace the null values

```python
ds_clean10=ds.replace(to_replace=np.nan, value= 5555)
```

```python
ds_clean10
```

Advantage of replace: Unlike fillna() we can replace any value with any value with replace function.

Task: Replace all the No with yesss in purchased column

```python
ds_clean11= ds.replace(to_replace='No', value='yesss')
```

```python
ds_clean11
```
